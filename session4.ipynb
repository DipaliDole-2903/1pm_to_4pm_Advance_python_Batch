# *args

```python
# sum of all values present in parameter
```

```python
def cal_sum(*args):
    sum=0
    for i in args:
        sum=sum+i # sum =0+2=2+3=5+4=9
    return f"sum of all values={sum}"
```

```python
cal_sum(2,3,4)
```

```python
cal_sum(4,5,2,5,6,23,5,7,9)
```

```python
lst=[4,5,6,7,10,39,49]
```

```python
# list unpacking
cal_sum(*lst)
```

```python
def restaurent_billing(**kwargs):
    bill = 0
    for food,qty in kwargs.items():
        amt=menu.get(food)*qty
        bill=bill+amt
        print(f"Food;{food},Quantity :{qty},amount : {amt}")
    return f"Total bill = {bill}"
```

```python
# generator 
# 1)yield
# 2)next
```

```python
def fib_number():
    a,b=0,1
    while True:
        yield a
        a,b=b,a+b
# a=0,b=1
# a,b=1,2
# 0,1,1
```

```python
result=fib_number()
```

```python
for i in range(5):
    print(next(result))
```
