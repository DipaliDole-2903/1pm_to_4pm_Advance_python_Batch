```python
def calculate_square(n):
    print(f"the square of number {n} is {n*n}")
```

```python
t2=threading.Thread(target=calculate_square,args=(5,))
t2.start()
```

```python
l1=[10,20,2,3,5,7,8,9,11,22]
```

```python
for i in l1:
    t2=threading.Thread(target=calculate_square,args=(i,))
    t2.start()
```

```python
import urllib.request   
def download_file(lnk):
    file_name=lnk.split("/")[-1]
    print("file is downloading.........")
    path="C:/Users/admin/Downloads/"+file_name
    urllib.request.urlretrieve(lnk,path)
    print(f"{file_name} is downloaded")
```

```python
download_file("https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/data.csv")
```

```python
lst=[
    "https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/data.csv",
"https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/test.csv",
"https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/train.csv"
]
```

```python
for i in lst:
    th=threading.Thread(target=download_file,args=(i,))
    th.start()
```
