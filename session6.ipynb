# Inheritance
- Accessing property and functions from parent class to child class or from one class to another class is claled as inheritance.
- there is no multiple inheritance in python like java

# single inheritance

```python
# Parent class
class Customer:
    def __init__(self,name,age,prod_id):
        self.name=name
        self.age=age
        self.prod_id=prod_id
    def display(self):
        print(f'Name : {self.name}\nAge : {self.age}\nProd_id : {self.prod_id}')
```

```python
# Child class
class product(Customer):
    def __init__(self,name,age,prod_id,product_name,price):
        super().__init__(name,age,prod_id) #super is used to access a parent class properties
        self.product_name=product_name
        self.price=price
    def show(self):
        super().display()
        print(f"product_name : {self.product_name}\nPrice : {self.price}")




```

# create an object == always create an object of child class

```python
p1=product('smith',24,'02587','keyboard','800rs')
```

```python
p1.show()
```

# multilevel inheritance

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def info(self):
        print(f'Name : {self.name}\nAge : {self.age} ')
```

```python
class Teacher(Person):
    def __init__(self,name,age,subject):
        super().__init__(name,age)
        self.subject=subject
    def information(self):
        super().info()
        print(f"Subject : {self.subject}")
```

```python
class departmentHead(Teacher):
    def __init__(self,name,age,subject,department):
        super().__init__(name,age,subject)
        self.department=department
    def display(self):
        super.info()
        print(f"\nDepartment : {self.department}")
```

```python
t1=departmentHead('vishal','28','Maths','Mechanical')
```

```python
t1.dispaly()
```

# Hierarchical inheritance

```python
class Parent:
    def __init__(self,name,plot):
        self.name=name
        self.plot=plot
```

```python
class child1(Parent):
    def __init__(self,name,plot,childname):
        super().__init__(name,plot)
        self.childname=childname
    def display(self):
        print(f"childname={self.childname}{self.name}\nPlot={self.plot}")
```

```python
c1=child1("mahesh mishra","15*20","vishal")
```

```python
c1.display()
```

```python
class child2(Parent):
    def __init__(self,name,plot,childname):
        super().__init__(name,plot)
        self.childname=childname
    def display(self):
        print(f"childname={self.childname}   {self.name}\nplot= {self.plot}")
```

```python
c2=child2('dipali','10*10','Anshul')
```

```python
c2.display()
```
