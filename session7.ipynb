- Two types in polymorphism
1) function overloading // not used in python
2) function overriding
3) operator overloading
4) operator overriding

```python
# function overriding
class India:
    def capital(self):
        print("New delhi is capital of India")
    def language(self):
        print("hindi is official language of India")
```

```python
class USA:
    def capital(self):
        print("Washington D.C is capital of India")
    def language(self):
        print("English is official language of India")
```

```python
# create an object
```

```python
I=India()
u=USA()
I.capital()
u.language()
```

```python
for i in I,u:
    i.capital()
    i.language()
```

# operator Overloading

- + = __add__
- - = __sub__
- * = __mul__
- / = __div__

- & = __and__
- | = __or__

- > = __gt__
- < = __lt__
- >= = __gte__
- <= = __lte__
- = = __eq__


```python
class Addition:
    def __init__(self,a):
        self.a=a
    def __add__(self,other):
        return self.a+other.a
```

```python
obj1=Addition(10)
obj2=Addition(20)
```

```python
obj1+obj2
```

```python
# greater than 
```

```python
class compare_age:
    def __init__(self):
        self.name=input("Enter your name")
        self.age=int(input("enter your age"))
    def __gt__(self,other):
        if self.age>other.age:
            return f"{self.name} is older than {other.name}"
        else:
            return f"{self.name} is younger than {other.name}"
```

```python
person1= compare_age()
```

```python
person2 = compare_age()
```

```python
person1 > person2
```

```python
class check_divisibility:
    def __init__(self):
        self.number=int(input("Enter a number"))
    def __and__(self,other):
        if self.number%2==0 and other.number%2==0:
            return "Both number is even"
        else:
            return "Both nubers are odd"
```

```python
o1=check_divisibility()
```

```python
o2=check_divisibility()
```

```python
o1 & o2
```

```python
# complex number = combination of real and imaginary number
# a+ib 
# a= 10,b=20
# a=4,b=5

# res = 14+i30
```

```python
class Complex_number:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __add__(self,other):
        return f"{self.a+other.a} + i{self.b+other.b}"
```

```python
c1= Complex_number(4,7)
```

```python
c2=Complex_number(9,3)
```

```python
c1+c2
```
