```python
# write a function in python calculate hypoteneus of a right angled triangle.
# (AC)^2=(AB)^2+(BC)^2
# C^2=a^2+b^3
# c=(a^2+b^2)^0.5
# 4,5
```

```python
# write a function in python which takes 3 sides and determines wherher 3 sides form a triangle or not.
```

```python
def cal_sides_triangle(a,b,c):
    if a + b > c and a + c > b and b + c > a:
        print('it is forming a triangle')
    else:
        print('it is not forming a triangle')
```

```python
cal_sides_triangle(4,5,6)
```

```python
# write a function in python that calculates area of triangle when 3 sides are given.
```

```python
def cal_area_triangle(a,b,c):
    s=(a+b+c)//2
    area=(s*(s-a)*(s-b)*(s-c))**0.5
    print(f'area of triangle is {area}')
```

```python
cal_area_triangle(4,5,6)
```
