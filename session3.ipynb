```python
def information(*args):
    print(f"values:{args}")
    print(f"type:{type(args)}")
    for i in args:
        print(i)
```

```python
# return multiplication of a number
# 2,4,6,3=144
```

```python
def return_multiplication(*args):
    m=1
    for i in args:
        m=m*i
    return m
```

```python
return_multiplication(2,4,5,6,3,2,5)
```

```python
lst=[3,4,5,6,7]
```

```python
return_multiplication(*lst)
# you need to unpack first by giving *
```

```python
return_multiplication(2,4,3,1)
```

```python
cart=[
    ('laptop',35000,2),
    ('keyboard',6000,2),
    ("mouse",200,3),
    ('mouse pad',100,5)
]
```

```python
def print_info(*args):
    total_price=0
    # print information
    for name,price,qty in args:
        print(f"name:{name},price:{price},quantity:{qty}")
        print("-----------------------------------------------")
        total_price=total_price+price
    return total_price
```

```python
print_info(*cart)
```

```python
cart=[
    ("smith",21,30000),
    ("alice",22,45000),
    ("gayatri",24,50000)
]
```

```python
def emp_info(*args):
    sal=0
    for name,age,salary in args:
        print(f"Name:{name},age:{age},salary:{salary}")
        sal=sal+salary
    return(f"final salary = {sal}")
```

```python
emp_info(*cart)
```

```python
# *args=act as a tuple
# **args = act as a dictionary
```

```python
def display(**kwargs): #kw for key and value pair me value stored
    print(f"values:{kwargs}")
    for key,value in kwargs.items():
        print(f"key={key}\nvalue={value}")
```

```python
display(name="dipali",job_role="student")
```
