```python
class Person:
    def properties(self,name,age,salary):  #self is used to defining a class properties
        self.name=name
        self.age=age
        self.salary=salary
    def display_data(self):
        print(f'Name={self.name}\nAge:{self.age}\nSalary{self.salary}')
```

# create an object

```python
person1=Person()
```

```python
person1.properties('john',25,50000) # at a time only one  object can be created
```

```python
person1.display_data()
```

```python
person2=Person()
```

```python
person2.properties("smith",24,60000)
```

```python
person2.display_data()
```

# create a class

```python
class Customer:
    def data(self,name,email,product_id):
        self.name = name
        self.email=email
        self.product_id=product_id
    def show_data(self):
        print(f'Customer Name:{self.name}\nEmail:{self.email}\nProduct_id:{self.product_id}')
```

# Bank Account Example

```python
class Bank_account:
    def __init__(self,ac_number,ac_holder_name,Balance):
        self.ac_number=ac_number
        self.ac_holder_name=ac_holder_name
        self.Balance=Balance
    def display(self):
        print(f"Account_No : {self.ac_number}\nAc_holder_name : {self.ac_holder_name}\nBalance : {self.Balance}")
    def deposit(self,deposit_money):
        self.deposit_money=deposit_money
        self.Balance=self.Balance+self.deposit_money
        print(f'The Money in account after deposit of {self.deposit_money} is {self.Balance}')
    def withdraw(self,withdraw_money):
        self.withdraw_money=withdraw_money
        if self.withdraw_money<=self.Balance:
            self.Balance=self.Balance-self.withdraw_money
            print(f'The Money in account after withdraw of {self.withdraw_money} is {self.Balance}')
        else:
            print("insufficient money in your bank account")
```

```python
p1=Bank_account('mahesh','147852',50000)
```

```python
p1.deposit(2000)
```

```python
p1.withdraw(10000)
```
